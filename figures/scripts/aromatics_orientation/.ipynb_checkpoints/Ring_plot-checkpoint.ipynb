```python
def colorbar_label_formatter(value, ax):
    if value <= 0.0002:
        return 'low'
    elif value >= 0.0018:
        return 'high'
    else:
        return None #f'{value:.5f}'

sc=['SCF', 'SCY', 'SCW']
def list_angles(sc):
    data1_0=[]
    data1_1=[]
    data1_2=[]
    data1_3=[]
    if sc=='SCW':
        traj_num=[2,3,4]
    else:
        traj_num=[1,2,3]
    for t in traj_num:
        for k in range(2,6):
            d=pd.read_table(coor_file.format(sc, t, 'angle_', k), delimiter=' ')
            d2=d[d['#z']<31]
            #Transfer the wanted angle value to a list
            data1_0.extend(d2['#z'].tolist())
            data1_1.extend(d2['theta1'].tolist())
            data1_2.extend(d2['theta2'].tolist())
            #data1_3.extend(d2['theta2_mc'].tolist())
    return data1_0, data1_1, data1_2, data1_3

#Repeat but extend the first created list if z is still the same

data1_0, data1_1, data1_2, data1_3=list_angles(sc[0])
data2_0, data2_1, data2_2, data2_3=list_angles(sc[1])
data3_0, data3_1, data3_2, data3_3=list_angles(sc[2])
print('datasets created')

# Creating subplots with multiple histograms
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(10, 5))
sns.set_style("white")

def colorbar_label_formatter(value, ax):
    if value <= 0.0002:
        return 'low'
    elif value >= 0.0019:
        return 'high'
    else:
        return None

sns.kdeplot(x=data1_0, y=data1_1, cmap="Reds", fill=True, ax=axes[0,0])
sns.kdeplot(x=data2_0, y=data2_1, cmap="Reds", fill=True, ax=axes[0,1])
sns.kdeplot(x=data3_0, y=data3_1, cmap="Reds", fill=True, ax=axes[0,2], cbar=True, cbar_kws=dict(format=colorbar_label_formatter))
print('line 1 done')

def colorbar_label_formatter(value, ax):
    if value <= 0.0002:
        return 'low'
    elif value >= 0.0013:
        return 'high'
    else:
        return None

sns.kdeplot(x=data1_0, y=data1_2, cmap="Blues", fill=True, ax=axes[1,0])
sns.kdeplot(x=data2_0, y=data2_2, cmap="Blues", fill=True, ax=axes[1,1])
sns.kdeplot(x=data3_0, y=data3_2, cmap="Blues", fill=True, ax=axes[1,2], cbar=True, cbar_kws=dict(format=colorbar_label_formatter))
print('line 2 done')

axes[0, 0].set_title('PHE')
axes[0, 1].set_title('TYR')
axes[0, 2].set_title('TRP')
print('3 done')


for i in [0,1]:
    for j in [0, 1, 2]:
        if i==1:
            axes[i,j].set_xlabel('z ($\AA$)')
        if j==0:
            axes[i,j].set_ylabel('Angles (°)')
        axes[i,j].set_xlim(0, 30)
        axes[i,j].set_ylim(0, 180)

# Adjusting layout for better spacing

#plt.tight_layout()
plt.savefig('angles_hist_{}.png'.format('ARO'), bbox_inches='tight')
# Display the figure
plt.show()
```
