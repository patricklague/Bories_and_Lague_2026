```python
#from Bio.PDB import *
#from Bio.PDB.DSSP import DSSP
#from Bio.PDB.DSSP import dssp_dict_from_pdb_file
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import math

coor_file='/media/bories/Backup/bories/Documents/Travail/results/homoPOPC-aa/homoPOPC-{}/analyses/traj{}/scripts/ring_coor/{}.dat'
```

```python
def normal_vector(x, y, z):
    points = np.array([x, y, z]).T
    # Vérifier si au moins 3 points sont fournis
    if len(points) < 3:
        raise ValueError("Au moins 3 points sont requis pour calculer un vecteur normal.")
    # Ajuster un modèle de régression linéaire aux points
    #model = LinearRegression()
    #model.fit(points[:, :-1], points[:, -1])
    X = np.array([x, y]).T
    Y = np.array(z)
    model = LinearRegression()
    model.fit(X,Y)
    # Obtenir le vecteur normal à partir des coefficients du modèle
    normal_vector = [model.coef_[0], model.coef_[1], -1]
    # Normaliser le vecteur normal
    normalized_normal_vector = normal_vector / np.linalg.norm(normal_vector)
    return normalized_normal_vector

def normal_vector2(x, y, z):
    points = np.array([x[0:3], y[0:3], z[0:3]]).T
    # Vérifier si au moins 3 points sont fournis
    if len(points) < 3:
        raise ValueError("Au moins 3 points sont requis pour calculer un vecteur normal.")
    # Centrage des points par rapport à l'origine
    centered_points = points - np.mean(points, axis=0)
    # Calculer le vecteur directeur 1 (entre le premier et le deuxième point)
    vector_1 = centered_points[1] - centered_points[0]
    # Calculer le vecteur directeur 2 (entre le premier et le troisième point)
    vector_2 = centered_points[2] - centered_points[0]
    # Calculer le produit vectoriel des vecteurs directeurs pour obtenir le vecteur normal
    normal_vector = np.cross(vector_1, vector_2)
    # Normaliser le vecteur normal
    normal_vector = normal_vector / np.linalg.norm(normal_vector)
    return normal_vector

def direction_vector(point1, point2):
    # Calculer le vecteur entre les deux points
    vector = [point2[0] - point1[0], point2[1] - point1[1], point2[2] - point1[2]]
    # Normaliser le vecteur
    normalized_vector = vector / np.linalg.norm(vector)
    return normalized_vector

def angle_between(v1, v2):
    #v1 et v2 doivent être des vecteurs unitaires
    #theta = np.arccos(v1[2]/v2[2])
    #theta=np.arccos(np.dot(v1, v2))
    theta=np.arccos(np.clip(np.dot(v1, v2), -1.01, 1.01))
    # Produit vectoriel
    #prod_vec = [(v1[1] * v2[2] - v1[2] * v2[1]),
    #                   -(v1[0] * v2[2] - v1[2] * v2[0]),
    #                   (v1[0] * v2[1] - v1[1] * v2[0])]
    #if math.isclose(prod_vec[0]**2+prod_vec[1]**2+prod_vec[2]**2, 0.0):
    #    return 0.0  # Vecteurs colinéaires, angle nul
    # Calcul de l'angle avec la tangente
    #tan_angle = prod_vec[0] / prod_vec[1]
    #angle = math.atan(tan_angle)
    #theta = angle
    return np.degrees(theta)
    #return v1[2]*v2[2]
    #return np.pi/2-theta

def angle_with_normal(structure, TM, v2):
    #calculate angle between the vector and the normal of the membrane
    x, y, z = get_ca_coord_part(structure, TM)
    v1=regressi_vector(x, y, z)
    return angle_between(v1, v2)
```

```python
scs = ['SCW']
trajs = [4, 2, 3]
carbons=['CD1', 'CD2', 'CE2', 'CZ2', 'CH2', 'CZ3', 'CE3']
Z_col=['z_CD1', 'z_CD2', 'z_CE2', 'z_CZ2', 'z_CH2', 'z_CZ3', 'z_CE3']
X_col=['x_CD1', 'x_CD2', 'x_CE2', 'x_CZ2', 'x_CH2', 'x_CZ3', 'x_CE3']
Y_col=['y_CD1', 'y_CD2', 'y_CE2', 'y_CZ2', 'y_CH2', 'y_CZ3', 'y_CE3']
all_xyz=X_col+Y_col+Z_col

def get_v2(row):
    return [0,0,row['z_temp']/abs(row['z_temp'])]

def add_normal(row):
    x1=[row['x_CD2'], row['x_CZ2'], row['x_CZ3'], row['x_CD1'], row['x_CE2'], row['x_CH2'], row['x_CE3']]
    y1=[row['y_CD2'], row['y_CZ2'], row['y_CZ3'], row['y_CD1'], row['y_CE2'], row['y_CH2'], row['y_CE3']]
    z1=[row['z_CD2'], row['z_CZ2'], row['z_CZ3'], row['z_CD1'], row['z_CE2'], row['z_CH2'], row['z_CE3']]
    return normal_vector2(x1, y1, z1)

def add_dir(row): #pour CG -> CZ
    point1=[row['x_CD2'], row['y_CD2'], row['z_CD2']]
    point2=[row['x_CH2'], row['y_CH2'], row['z_CH2']]
    return direction_vector(point1, point2)

def add_dir_mc(row): #pour CG -> CE1
    point1=[row['x_CD1'], row['y_CD1'], row['z_CD1']]
    point2=[row['x_CH2'], row['y_CH2'], row['z_CH2']]
    return direction_vector(point1, point2)

def add_angle1(row):
    v1=row['v1']
    v2=row['v2']
    return angle_between(v1, v2)
    
def add_angle2(row):
    v3=row['v3']
    v2=row['v2']
    return angle_between(v3, v2)

def add_angle3(row):
    v4=row['v4']
    v2=row['v2']
    return angle_between(v4, v2)


for sc in scs:
    for t in trajs:
        for k in range(1,4):
            for j in range(len(carbons)):
                c=carbons[j]
                if j==0:
                    d=pd.read_table(coor_file.format(sc, t, f'{c}_coor_mono_{k}'), delimiter=' ')
                    d.rename(columns={'x': 'x_{}'.format(c), 'y': 'y_{}'.format(c), 'z': 'z_{}'.format(c)}, inplace=True)
                    d.drop(['#frame', 'C_type', 'index', 'resid'], axis=1, inplace=True)
                else:
                    temp=pd.read_table(coor_file.format(sc, t, f'{c}_coor_mono_{k}'), delimiter=' ')
                    d['x_{}'.format(c)], d['y_{}'.format(c)], d['z_{}'.format(c)]=temp['x'], temp['y'], temp['z']
                    del temp
            #display(d)
            d['z_temp']=d[Z_col].mean(axis=1)
            d['v2']=d.apply(get_v2, axis=1)
            d['z_mean']=d['z_temp']*(d['z_temp']/abs(d['z_temp']))
            d.drop(['z_temp'], axis=1, inplace=True)               
            d['v1']=d.apply(add_normal, axis=1)
            d['v3']=d.apply(add_dir, axis=1)
            d['v4']=d.apply(add_dir_mc, axis=1)
            d.drop(all_xyz, axis=1, inplace=True)
            #display(d)
            d['angle1']=d.apply(add_angle1, axis=1)
            d['angle2']=d.apply(add_angle2, axis=1)
            d['angle3']=d.apply(add_angle3, axis=1)
            d.drop(['v1', 'v2', 'v3', 'v4'], axis=1, inplace=True)
            #display(d)
            np.savetxt(coor_file.format(sc, t, f'angle_mono_{k}'), d, fmt='%.3f %.6f %.6f %.6f', delimiter=' ', header='z theta1 theta2 theta2_mc', comments='#')
            print('angle_mono_{}.png has been saved...'.format(k))
print('C\'est bon, j\'ai fini !!!')
```
