# Extraction des données exactes depuis l'image

Extraction manuelle des coordonnées (x, y) pour chaque courbe d'acide aminé.
- Centre: z = 0 (centre de la membrane)
- Zone jaune (hydrophobe): -20 à +20 Å (40 Å de large)
- Zone violette (interface): environ -25 à -20 Å et +20 à +25 Å
- Zone bleue (eau): au-delà de ±25 Å

```python
import numpy as np
import matplotlib.pyplot as plt

# Liste des positions z (centres des bins) - COMMUNE À TOUS LES RÉSIDUS
# De -50 à 50 par pas de 4 Å
z_positions = np.arange(-48, 50, 4)  # [-48, -44, -40, ..., 44, 48]
print(f"Positions z: {z_positions}")
print(f"Nombre de bins: {len(z_positions)}")

# Fonction pour parser les données copiées
def parse_data(data_string):
    """
    Parse les données au format 'Bar0, valeur' et retourne un array numpy
    Vous pouvez coller directement vos données entre les triple quotes
    """
    lines = data_string.strip().split('\n')
    values = []
    for line in lines:
        if ',' in line:
            # Extrait la valeur après la virgule
            value = float(line.split(',')[1].strip())
            values.append(value)
    return np.array(values)
```

```python
# Charger les données depuis le fichier sauvegardé et les distributions mono
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
import os

# Charger les données MPM
mpm_data = pd.read_csv('../data/mpm_freq.dat', sep='\t')
print("Données MPM chargées:")
print(mpm_data.head())

# Mapping des codes SC vers les noms d'acides aminés
sc_to_aa = {
    'sca': 'ALA', 'scc': 'CYS', 'scd': 'ASP', 'sce': 'GLU',
    'scf': 'PHE', 'glyd': 'GLY', 'schd': 'HIS', 'sci': 'ILE',
    'sck': 'LYS', 'scl': 'LEU', 'scm': 'MET', 'scn': 'ASN',
    'scp': 'PRO', 'scq': 'GLN', 'scr': 'ARG', 'scs': 'SER',
    'sct': 'THR', 'scv': 'VAL', 'scw': 'TRP', 'scy': 'TYR'
}

# Charger les distributions mono
mono_distributions = {}
base_path = '../data/mono_data/popc-mono/distributions'

for sc_code, aa_name in sc_to_aa.items():
    folder_path = os.path.join(base_path, sc_code)
    if os.path.exists(folder_path):
        # Charger tous les fichiers trajectory*.dat et faire la moyenne
        traj_files = glob.glob(os.path.join(folder_path, 'trajectory*.dat'))
        if traj_files:
            all_data = []
            for f in traj_files:
                df = pd.read_csv(f, sep='\t', header=None, names=['z', 'prob', 'std'])
                all_data.append(df)
            
            # Moyenne des trajectoires
            combined = pd.concat(all_data)
            avg_data = combined.groupby('z')['prob'].mean().reset_index()
            mono_distributions[aa_name] = avg_data
            print(f"✓ {aa_name} ({sc_code}): {len(traj_files)} trajectoires, {len(avg_data)} points")

print(f"\nTotal: {len(mono_distributions)} distributions mono chargées")
```

```python
# Créer le graphique comparatif MPM vs Mono
fig, axes = plt.subplots(2, 4, figsize=(16, 8), sharey=True)
fig.subplots_adjust(wspace=0.05, hspace=0.3)

# Définition des couleurs (basées sur l'image)
colors = {
    'ASP': '#2ca02c', 'GLU': '#1f77b4',
    'ARG': '#2ca02c', 'LYS': '#1f77b4',
    'ASN': '#2ca02c', 'GLN': '#1f77b4', 'HIS': '#d62728',
    'GLY': '#2ca02c', 'PRO': '#1f77b4', 'SER': '#d62728', 'THR': '#bcbd22',
    'TRP': '#2ca02c', 'TYR': '#1f77b4',
    'ALA': '#2ca02c', 'MET': '#1f77b4',
    'CYS': '#2ca02c', 'VAL': '#1f77b4',
    'ILE': '#2ca02c', 'LEU': '#1f77b4', 'PHE': '#d62728'
}

def add_background_regions(ax):
    """Ajoute les régions colorées de fond"""
    # Région eau (bleu clair)
    ax.axvspan(-50, -20, alpha=0.15, color='#B8D4E8')
    ax.axvspan(20, 50, alpha=0.15, color='#B8D4E8')
    
    # Noyau hydrophobe (jaune)
    ax.axvspan(-20, 20, alpha=0.2, color='#FFFFCC')
    
    # Lignes verticales pointillées aux limites
    ax.axvline(-20, color='gray', linestyle='--', linewidth=0.8, alpha=0.5)
    ax.axvline(20, color='gray', linestyle='--', linewidth=0.8, alpha=0.5)

def normalize_distribution_in_range(z_values, prob_values, z_min=-50, z_max=50):
    """
    Normalise une distribution pour que l'intégrale entre z_min et z_max = 1
    Prend en compte la largeur des bins (différente pour MPM et Mono)
    """
    # Filtrer les valeurs dans la plage
    mask = (z_values >= z_min) & (z_values <= z_max)
    prob_in_range = prob_values[mask]
    z_in_range = z_values[mask]
    
    if len(z_in_range) < 2:
        return z_in_range, prob_in_range
    
    # Calculer la largeur des bins (on suppose qu'elle est constante)
    dz = np.abs(z_in_range[1] - z_in_range[0])
    
    # Normaliser en tenant compte de la largeur des bins
    # Intégrale ≈ somme(prob_i × dz)
    integral = np.sum(prob_in_range) * dz
    
    if integral > 0:
        prob_normalized = prob_in_range / integral
    else:
        prob_normalized = prob_in_range
    
    return z_in_range, prob_normalized

# Organisation des panneaux
panels = [
    # Ligne 1
    {'title': 'Negative', 'amino_acids': ['ASP', 'GLU'], 'pos': (0, 0)},
    {'title': 'Positive', 'amino_acids': ['ARG', 'LYS'], 'pos': (0, 1)},
    {'title': 'Hydrophilic polar', 'amino_acids': ['ASN', 'GLN', 'HIS'], 'pos': (0, 2)},
    {'title': 'Low weight', 'amino_acids': ['GLY', 'PRO', 'SER', 'THR'], 'pos': (0, 3)},
    # Ligne 2
    {'title': 'Aromatic', 'amino_acids': ['TRP', 'TYR'], 'pos': (1, 0)},
    {'title': 'Mildly hydrophobic', 'amino_acids': ['ALA', 'MET'], 'pos': (1, 1)},
    {'title': 'Low-weight hydrophobic', 'amino_acids': ['CYS', 'VAL'], 'pos': (1, 2)},
    {'title': 'Hydrophobic', 'amino_acids': ['ILE', 'LEU', 'PHE'], 'pos': (1, 3)},
]

# Tracer chaque panneau
for panel in panels:
    row, col = panel['pos']
    ax = axes[row, col]
    
    # Ajouter les régions de fond
    add_background_regions(ax)
    
    # Tracer chaque acide aminé
    for aa_name in panel['amino_acids']:
        color = colors[aa_name]
        
        # Courbe MPM (ligne pleine) - avec normalisation tenant compte de dz=4Å
        if aa_name in mpm_data.columns:
            z_norm_mpm, prob_norm_mpm = normalize_distribution_in_range(
                mpm_data['z'].values, mpm_data[aa_name].values, z_min=-50, z_max=50)
            ax.plot(z_norm_mpm, prob_norm_mpm, 
                   label=f'{aa_name} (MPM)', color=color, linewidth=1.5, linestyle='-')
        
        # Courbe Mono (ligne pointillée) - avec normalisation tenant compte de dz=1Å
        if aa_name in mono_distributions:
            mono_df = mono_distributions[aa_name]
            z_norm, prob_norm = normalize_distribution_in_range(
                mono_df['z'].values, mono_df['prob'].values, z_min=-50, z_max=50)
            ax.plot(z_norm, prob_norm, 
                   label=f'{aa_name} (Mono)', color=color, linewidth=1.5, linestyle='--')
    
    # Mise en forme
    ax.set_ylim(0, 0.07)
    ax.set_xlim(-50, 50)
    ax.set_title(panel['title'], fontsize=11, fontweight='bold', loc='left')
    ax.legend(loc='upper right', frameon=False, fontsize=7, ncol=2)
    ax.set_xlabel('Position (Å)', fontsize=10)
    
    # Étiquette Y seulement pour les panneaux de gauche
    if col == 0:
        ax.set_ylabel('Probability density (Å$^{-1}$)', fontsize=10)
    
    # Paramètres des ticks
    ax.tick_params(labelsize=9)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('../plot/comparison_mpm_mono.png', dpi=300, bbox_inches='tight')
plt.show()

print("✓ Graphique sauvegardé dans ../plot/comparison_mpm_mono.png")
print("Note: Les distributions sont normalisées en densité de probabilité (intégrale = 1)")
```

## Explications sur les données

**Structure de la membrane:**
- **Centre (z = 0)**: Centre de la membrane bicouche lipidique
- **Zone jaune (-20 à +20 Å)**: Noyau hydrophobe de 40 Å de largeur (chaînes acyles des lipides)
- **Zone violette/rose (-25 à -20 et +20 à +25 Å)**: Interface membrane-eau (têtes polaires des lipides)
- **Zone bleue (au-delà de ±25 Å)**: Phase aqueuse (bulk water)

**Organisation des acides aminés:**
- **Chargés** (ASP, GLU, ARG, LYS): Pics dans l'interface, évitent le noyau hydrophobe
- **Polaires hydrophiles** (ASN, GLN, HIS): Préfèrent l'interface
- **Aromatiques** (TRP, TYR): Fort pic dans l'interface, positionnement spécifique
- **Hydrophobes** (ILE, LEU, PHE): Pics centraux dans le noyau hydrophobe
